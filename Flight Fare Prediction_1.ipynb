<h1> DATE OF JOURNEY </h1>

```python
df_data['Date_of_Journey'] = pd.to_datetime(df_data['Date_of_Journey'])
```

<h1> DURATION </h1>

```python
def to_minutes(duration):
    if ('h' in duration) and ('m' in duration):
        time = duration.split()
        hours = int(time[0][:-1])
        minutes = int(time[1][:-1])
        return hours*60 + minutes
    else:
        time = duration.split()
        hours = int(time[0][:-1])
        return hours*60
    
```

```python
df_data['Duration'] = df_data['Duration'].apply(to_minutes)
df_data.head()
```

```python
df_data['Dep_Time_Hrs'] = df_data['Dep_Time'].dt.hour
df_data['Dep_Time_Mins'] = df_data['Dep_Time'].dt.minute
df_data['Arrival_Time_Hrs'] = df_data['Arrival_Time'].dt.hour
df_data['Arrival_Time_Mins'] = df_data['Arrival_Time'].dt.minute
```
